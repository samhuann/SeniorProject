```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")

# Simulate data from a bivariate Gaussian
n = 10000
mean = [0, 0]
cov = [(2, .4), (.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T

# Draw a combo histogram and scatterplot with density contours
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to prune the dataframe by keeping X values within a specified range
def prune_by_range(data, x_column, x_min, x_max):
    pruned_data = data[(data[x_column] >= x_min) & (data[x_column] <= x_max)]
    return pruned_data

# Function to prune the dataframe by averaging every K rows after sorting by X values
def prune_by_average(data, x_column, k):
    sorted_data = data.sort_values(by=[x_column])  # Ensure x_column is passed as a list
    num_rows = len(sorted_data)
    num_groups = num_rows // k
    pruned_data = pd.DataFrame()
    for i in range(num_groups):
        start_idx = i * k
        end_idx = min((i + 1) * k, num_rows)
        group_data = sorted_data.iloc[start_idx:end_idx]
        averaged_row = group_data.mean(axis=0)
        pruned_data = pruned_data.append(averaged_row, ignore_index=True)
    return pruned_data

# Load large table (replace 'large_table.xlsx' with your file)
data = pd.read_excel("Samuel_Project/SampleData.xlsx")

# Print columns and prompt user to choose X column
print("Columns in the dataframe:")
print(data.columns)
x_column = input("Enter the name of the column containing X values: ")

# Ask the user what kind of pruning they want to do
pruning_method = input("Choose pruning method (Range/Average): ").capitalize()

# Perform pruning based on user selection
if pruning_method == "Range":
    x_min = float(input("Enter minimum X value: "))
    x_max = float(input("Enter maximum X value: "))
    pruned_data = prune_by_range(data, x_column, x_min, x_max)
elif pruning_method == "Average":
    k = int(input("Enter number of rows to average (K): "))
    pruned_data = prune_by_average(data, x_column, k)

# Print the pruned dataframe
print("Pruned Data:")
print(pruned_data)

# Ask if user wants to generate a new graph
generate_graph = input("Generate a new graph (yes/no)? ").lower()

if generate_graph == "yes":
    # Plot the pruned data
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=pruned_data, x=x_column, y='Y')
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Pruned Data")
    plt.show()
```
