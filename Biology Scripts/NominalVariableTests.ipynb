```python
# Example data
num_success = 3  # Number of successes
num_trials = 15   # Number of trials
null_prob = 0.1# Null hypothesis probability
alternative = 'two-sided'

# Perform the binomial test
p_value = binomtest(num_success, num_trials, null_prob, alternative)

# Output the p-value
p_value.pvalue
```
