```python
dfV = df
dfV.columns = dfV.columns.map(str)
posMean = dfV["0"].mean()
dmsoNegMean = dfV["DMSO NEG"].mean()
for c in dfV.columns:
    dfV[c] = 100*((dfV[c]-dmsoNegMean)/(posMean-dmsoNegMean)) 
dfV1 = dfV.loc[:,"0":"512.1"]
#loc: [rows,columns]
normalizedViability = dfV1.mean()
normalizedFrame = normalizedViability.to_frame(name="Normalized Viability")
normalizedFrame = normalizedFrame.rename_axis("Concentration")
print(normalizedFrame)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style
sns.set(style="darkgrid")

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(x='Concentration', y='Normalized Viability', data=normalizedFrame, marker='o', color='b', linewidth=2.5)

# Adding titles and labels
plt.title('MTT Assay', fontsize=16)
plt.xlabel('Concentration', fontsize=14)
plt.ylabel('Normalized Viability', fontsize=14)

# Show plot
plt.show()
```
