```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to transpose the dataframe
def transpose_dataframe(data):
    transposed_data = data.transpose()
    return transposed_data

# Load Excel dataframe (replace 'input_data.xlsx' with your Excel file)
data = pd.read_excel("Samuel_Project/sample_input.xlsx")

# Transpose the dataframe
transposed_data = transpose_dataframe(data)

# Print the transposed dataframe
print("Transposed Data:")
print(transposed_data)

# Ask user if they want to create a new graph
create_graph = input("Do you want to create a new graph (yes/no)? ").lower()

if create_graph == "yes":
    # Plot the transposed data
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=transposed_data.T, dashes=False)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Transposed Data")
    plt.legend(title="Data Sets", labels=transposed_data.index)
    plt.show()
```
